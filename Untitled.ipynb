```python
x=df.drop('Made Donation in March 2007',axis=1)
```

```python
y = df['Made Donation in March 2007']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=66)
```

```python
df["Made Donation in March 2007"].value_counts()
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE()
```

```python
sm
```

```python
x_train_smote,y_train_smote = sm.fit_resample(x_train, y_train)
```

```python
from collections import Counter
```

```python
print("Before applying Balancing Technique",Counter(y_train))
print("After applying Balancing Technique",Counter(y_train_smote))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr
```

```python
lr.fit(x_train_smote,y_train_smote)
```

```python
pred = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,pred)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,pred)
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC()
svm
```

```python
svm.fit(x_train_smote,y_train_smote)
```

```python
x_test
```

```python
y_pred = svm.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,f1_score,classification_report
```
